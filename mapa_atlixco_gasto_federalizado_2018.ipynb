```python
import folium
```

```python
mapa_atlixco = folium.Map (
    location=[18.912500, -98.427778],
    zoom_start=13
)
```

```python
mapa_atlixco
```

```python
for i in range(len(etiquetas)):
  folium.Marker(lat_long[i],
                popup=f"Poblacion beneficiada: {beneficiarios[i][0]} personas", 
                
                tooltip=etiquetas[i][0],
                icon=folium.Icon(color='gray', prefix='fa', icon='check')
  ).add_to(mapa_atlixco)
```
